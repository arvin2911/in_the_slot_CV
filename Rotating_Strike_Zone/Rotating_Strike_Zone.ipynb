```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
import numpy as np
from math import *
```

```python
top = int(input())
low = int(input())

def draw_baseball_home_plate_3d(top, low):
    # Define the vertices of a baseball home plate in 3D
    vertices_base = np.array([
        [0, 0, low],
        [6 * np.sqrt(2), 6 * np.sqrt(2), low],
        [6 * np.sqrt(2), 6 * np.sqrt(2) + 8.5, low],
        [-6 * np.sqrt(2), 6 * np.sqrt(2) + 8.5, low],
        [-6 * np.sqrt(2), 6 * np.sqrt(2), low],
    ])

    # Extrude the pentagon to create the sides
    vertices_top = vertices_base + [0, 0, (top-low)]
    
    # Combine base and top vertices
    vertices = np.vstack([vertices_base, vertices_top])

    # Create the faces of the prism
    faces = [
        [vertices_base[i], vertices_base[(i + 1) % 5], vertices_top[(i + 1) % 5], vertices_top[i]] for i in range(5)
    ]
    faces.append(vertices_base)
    faces.append(vertices_top)

    # Create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the prism
    ax.add_collection3d(Poly3DCollection(faces, edgecolor='black', facecolor='brown', alpha=0.6))

    # Set axis limits
    ax.set_xlim([-20, 20])
    ax.set_ylim([-20, 20])
    ax.set_zlim([0, top + 10])

    # Display the plot
    plt.show()

# Call the function to draw the 3D baseball home plate graph
draw_baseball_home_plate_3d(top, low)
```
