```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def plot_pentagonal_prism(ax, side_length, height):
    # Define the vertices of the pentagon base
    base_vertices = np.array([
        [side_length * 0.5 * np.cos(2 * np.pi * i / 5), side_length * 0.5 * np.sin(2 * np.pi * i / 5)]
        for i in range(5)
    ])

    # Extrude the base vertices to create the prism
    vertices = np.array([
        [*vertex, 0] for vertex in base_vertices
    ])
    vertices = np.vstack([vertices, vertices[0]])  # Close the base

    extruded_vertices = vertices.copy()
    extruded_vertices[:, -1] = height
    vertices = np.vstack([vertices, extruded_vertices])

    # Create a Poly3DCollection and add it to the plot with shading
    poly3d_collection = Poly3DCollection([vertices], edgecolor='black', facecolors='cyan', alpha=0.5)
    ax.add_collection3d(poly3d_collection)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Set the side length and height of the pentagonal prism
side_length = 1.0
height = .5

# Call the function to plot the pentagonal prism
plot_pentagonal_prism(ax, side_length, height)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
